```python
import pandas as pd

data2.dropna(axis=0, how="all", inplace=True)

def automate(value):
    formats = ["%m/%d/%y %H:%M", "%m-%d-%Y %H:%M:%S", "%m-%d-%Y"]
    result_format = "%d-%m-%y"
    for dt_format in formats:
        try:
            dt_obj = pd.to_datetime(value, format=dt_format, errors="coerce")
            if not pd.isna(dt_obj):
                return dt_obj.strftime(result_format)
        except ValueError:
            pass
    return pd.NaT

data2["Order Date"] = data2["Order Date"].apply(automate)
```

# Q4 What products are most often sold together ?

```python
df=data2[data2["Order ID"].duplicated(keep=False)]
df["Grouped"]=df.groupby("Order ID")["Product"].transform(lambda x: ",".join(x))
df.head(3)
```

```python
df=df[df[["Order ID","Grouped"]].duplicated(keep="first")][["Order ID","Grouped"]]
df.head(5)
```

```python
from itertools import combinations
from collections import Counter

count=Counter()
for row in df["Grouped"]:
    row_list=row.split(",")
    count.update(Counter(combinations(row_list,2)))
#print(count)
count.most_common()
```

```python
df.head(3)
```

```python
# MY MY MY APPROACH

from itertools import combinations
emp=[]
for row in df["Grouped"]:
    row_list=row.split(",")
    emp.extend(combinations(row_list,2))
df1=pd.Series(emp).value_counts()
df1
```
